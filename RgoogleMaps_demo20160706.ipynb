```r
require("RgoogleMaps")
```

```r
datag <- read.csv("http://sutchy.cocolog-nifty.com/sutchy/files/sample_geo20160616.csv",fileEncoding='cp932')

```

```r
head(datag)
```

```r
map = GetMap(c(lat = 36.0, lon = 136.0), # 地図の中心点を指定。緯度経度
zoom=5, # 地図の倍率指定
size = c(465, 640),
 sensor="false", 
 hl="ja", # 地図表示の言語を指定。
maptype= "terrain" #("roadmap", "mobile", "satellite", "terrain", "hybrid", "mapmaker-roadmap", "mapmaker-hybrid")
 ,path = "&style=feature:all|element:labels|visibility:off&style=feature:all|element:all|saturation:-50|lightness:60"
 #,GRAYSCALE = T
)
PlotOnStaticMap(map,lat=datag$fY,lon=datag$fX,FUN=points, cex=datag[,5]/1000, col="#00266F60", pch=19) 
```
